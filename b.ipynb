```python
import requests
import json
import pandas as pd
import time
import os
```

```python
df = pd.read_csv('anime.csv')
```

```python
# get mal_id from the dataframe
mal_id = df['mal_id'].tolist()
print(mal_id)
```

```python
# count the mal_id
count = len(mal_id)
print(f"Total mal_id: {count}")
```

```python
# File untuk simpan checkpoint
checkpoint_file = 'checkpoint.json'
output_file = 'anime_full_data.json'
```

```python
# Mulai dari index terakhir jika file checkpoint ada
start_index = 0
anime_data = []
```

```python
if os.path.exists(checkpoint_file):
    with open(checkpoint_file, 'r', encoding='utf-8') as f:
        checkpoint = json.load(f)
        start_index = checkpoint.get("last_index", 0)
        anime_data = checkpoint.get("data", [])
    print(f"Melanjutkan dari index {start_index}...")

print(f"Total anime: {len(mal_id)}")
```

```python
# Mulai pengambilan data
for i in range(start_index, len(mal_id)):
    anime_id = mal_id[i]
    url = f"https://api.jikan.moe/v4/anime/{anime_id}/full"

    try:
        response = requests.get(url)
        if response.status_code == 429:
            print("Rate limit reached. Sleeping for 60 seconds...")
            time.sleep(60)
            continue

        data = response.json()
        anime_data.append(data)

        print(f"[{i+1}/{len(mal_id)}] Data for ID {anime_id} retrieved.")
        time.sleep(1)

        # Simpan checkpoint tiap 100 data
        if (i + 1) % 100 == 0 or i == len(mal_id) - 1:
            with open(checkpoint_file, 'w', encoding='utf-8') as f:
                json.dump({"last_index": i + 1, "data": anime_data}, f, ensure_ascii=False, indent=4)
            print(f"Checkpoint saved at index {i+1}")

    except Exception as e:
        print(f"[{i+1}] Error at ID {anime_id}: {e}")
        continue
# Simpan hasil akhir
with open(output_file, 'w', encoding='utf-8') as f:
    json.dump(anime_data, f, ensure_ascii=False, indent=4)

print(f"✅ Selesai. Data disimpan di: {output_file}")
```

```python
for i in range(count):
    anime_id = mal_id[i]  # get the mal_id for each iteration
    url = f"https://api.jikan.moe/v4/anime/{anime_id}/full"

    try:
        response = requests.get(url)
        data = response.json()
        print(f"Data for anime ID {anime_id} retrieved successfully.")
        time.sleep(0.5)  # to avoid hitting the API rate limit
    except Exception as e:
        print(f"Error retrieving data for anime ID {anime_id}: {e}")
        continue

```

```python
anime_id = 20 # contoh ID anime, ganti dengan ID yang diinginkan
url = f"https://api.jikan.moe/v4/anime/{anime_id}/full"

try:
    response = requests.get(url)
    response.raise_for_status()  # Raise HTTPError for bad responses (4xx or 5xx)

    data = response.json()
    print(json.dumps(data, indent=4)) # Menampilkan data dalam format yang lebih mudah dibaca

except requests.exceptions.RequestException as e:
    print(f"Error during request: {e}")
except json.JSONDecodeError as e:
    print(f"Error decoding JSON: {e}")
```

```python
# simpan data ke file JSON
with open('anime_data.json', 'w') as f:
    json.dump(data, f, indent=4)
# simpan data ke file JSON
```
