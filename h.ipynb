```python
from google_play_scraper import Sort, reviews_all

result = reviews_all(
    'com.shopee.id',
    sleep_milliseconds=0, # defaults to 0
    lang='id', # defaults to 'en'
    country='id', # defaults to 'us'
    sort=Sort.MOST_RELEVANT, # defaults to Sort.MOST_RELEVANT
    filter_score_with=None, # defaults to None
    count=10000, # defaults to 10000
)
```

```python
import pandas as pd
df = pd.DataFrame(result)
```

```python
df.to_csv('shopee_reviews.csv', index=False)
```

```python
df = pd.read_csv('shopee_reviews.csv')

df.head()
```

```python
# Ambil hanya kolom yang relevan
df.drop(columns=['reviewId', 'userName','userImage','thumbsUpCount', 'reviewCreatedVersion', 'at', 'replyContent', 'repliedAt', 'appVersion'], inplace=True)
```

```python
df.head()
```

```python
# Bersihkan teks ulasan
import re
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

factory = StemmerFactory()
stemmer = factory.create_stemmer()

def clean_text(text):
    text = text.lower()
    text = re.sub(r'http\S+', '', text)               # hapus link
    text = re.sub(r'[^a-zA-Z\s]', '', text)           # hapus angka & simbol
    text = re.sub(r'\s+', ' ', text).strip()          # hapus spasi berlebih
    text = stemmer.stem(text)                         # stemming bahasa Indonesia
    return text
```

```python
# Terapkan pembersihan
df['clean_review'] = df['content'].apply(clean_text)
```
